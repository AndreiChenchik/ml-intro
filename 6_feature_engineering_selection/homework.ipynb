### Simple version (1 point)

```python
# 1 point
def scale(X):
    X_mean = np.mean(X);
    X_std = np.std(X);

    return (X - X_mean) / X_std
```

```python
assert np.allclose(np.array(scale(wine_data)), StandardScaler().fit_transform(wine_data))
```

### Pipeline Version (2 points)

```python
# 2 points
from sklearn.base import BaseEstimator, TransformerMixin
class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, copy=True, with_mean=True, with_std=True):
        self.copy = copy
        self.with_mean=True
        self.with_std=True

    def fit(self, X, y=None):
        data = np.array(X)
        self.mean = np.mean(data.T, axis=-1) if self.with_mean else 0.0
        self.std = np.std(data.T, axis=-1) if self.with_std else 1.0
        return self

    def transform(self, X, y=None, copy=None):
        X = np.copy(X) if self.copy else X

        X = X.astype('float', copy=False)

        X -= self.mean
        X /= self.std

        return X        
            
```

```python
assert np.allclose(CustomScaler().fit_transform(wine_data), StandardScaler().fit_transform(wine_data))
```

## Exercise  3 - Preprocessing Pipeline (2 points)

```python
wine_train, wine_val, wine_labels_train, wine_labels_val = train_test_split(wine_data, wine_labels, 
                                                                            test_size=0.3, random_state=42)
```

Train model (for example, `LogisticRegression(solver='liblinear', penalty='l1')` on raw `wine_train` data; then train same model after data scaling; then add feature selection (and train model again on scaled data). For each experiment all required preprocessing steps (if any) should be wrapped into sklearn pipeline.

Measure `accuracy` of all 3 approaches on `wine_val` dataset. Describe and explain results.

```python
from sklearn.base import ClassifierMixin

def make_classifier_pipeline(
    classifier: ClassifierMixin = LogisticRegression(solver='liblinear', penalty='l1'),
    with_scaling: bool = False, with_feature_selection: bool = False
) -> Pipeline:
    steps = []

    if with_scaling:
        scaler = StandardScaler()
        steps.append(scaler)
    
    if with_feature_selection:
        selection_model = RandomForestClassifier(random_state=42)
        selector = SelectFromModel(selection_model)
        steps.append(selector)

    steps.append(classifier)

    return make_pipeline(*steps)
    

def get_clf_accuracy(clf, X_train = wine_train, y_train = wine_labels_train, 
                     X_test = wine_val, y_test = wine_labels_val) -> float:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    return accuracy_score(y_test, y_pred)


pipeline_base = make_classifier_pipeline()
pipeline_scaled = make_classifier_pipeline(with_scaling=True)
pipeline_scaled_filtered = make_classifier_pipeline(with_scaling=True, with_feature_selection=True)

print('Accuracy with scaling and feature selection:\n')
print('Result on original data: {:f}'.format(get_clf_accuracy(pipeline_base)))
print('Result after scaling {:f}'.format(get_clf_accuracy(pipeline_scaled)))
print('Result after scaling and feature selection {:f}'.format(get_clf_accuracy(pipeline_scaled_filtered)))
```

> There is no difference between training LogReg on original data vs training on scaled data, I assume that this is because we are not depended to features with low scale. Feature selection on the other hand works pretty damn well and removes the noise features!

## Exercise 4 - manual PCA (5 points)
The task is to solve PCA as an optimization problem, without explicitly doing eigen value decomposition.
In the most general setting PCA is minimization of reconstruction error of a projection of given rank $q$

$$\min_{\mu, \lambda_1,\ldots, \lambda_n, \mathbf{V}_q} \sum_{i=1}^n ||x_i - \mu - \mathbf{V}_q \lambda_i||^2$$

With a number of steps that can be found here https://stats.stackexchange.com/a/10260 this task transforms to
 $$\max_{u_i} \sum_{i=1}^q u_i^T \mathbf{S} u_i$$
 where $\mathbf{S}$ is the sample covariance matrix (after standartization) and $u_1, \ldots, u_q$ are the $q$ are orthonormal columns in $\mathbf{V}_q$.
 Let us solve this optimization problem with `scipy.optimize` library.
 
 Additional 2 point are given for visualization of the results.

### PCA (3 points)

```python
wine_data, wine_labels = wine_sklearn['data'], wine_sklearn['target']
```

Find a covariance matrix of standartized data and assing it to S.

```python
data = StandardScaler().fit_transform(wine_data)

S = np.cov(data, rowvar=False)
```

If your code is correct, the following assert should be Ok.

```python
assert np.allclose(np.linalg.norm(S), 5.787241159764733)
```

```python
from scipy.optimize import minimize

def objective(x):
    # your code: write objective of the problem (don't forget that scipy does min while we need max)
    return -(x.T@S@x)

def norm_constraint(x):
    # your code: constaraint norm of x to be 1, function should return 0 if constraint holds
    return np.linalg.norm(x) - 1 

con1 = {'type': 'eq', 'fun': norm_constraint}

x0 = np.ones(data.shape[1])

sol = minimize(objective, 
               x0, 
               constraints = [con1]
              )
x0 = sol.x
```

Hurray! We have first vector! Let's do another one.

```python
def orthogonality_constraint(x):
    # your code: x should be orthogonal to x0, function should return 0 if constraint holds
    return x.T@x0


con2 = {'type': 'eq', 'fun': orthogonality_constraint}

x1 = np.ones(data.shape[1])


sol = minimize(objective, 
               x1, 
               constraints = [con1, con2]
              )

x1 = sol.x
```

If your solution is correct, the following asserts should be Ok.

```python
assert np.allclose(x0@S@x0, 4.732436977583595)
assert np.allclose(x1@S@x1, 2.5110809296451233)
```

### Visualization (2 points)

Visualize the points after applying custom dimension reduction with 2 components.

```python
plot_scatter(data@x0, data@x1, auto_scaled=True , clusters=wine_labels, title='PCA with 2 components')
```

Let us classify handwritten digits 0, 1 and 2. 
To make task not so easy the images are binarized (no shadows of gray present) as it happens with xerocopied documents.

Let us also find out to which parts of an image there's no need to look in order to clasify three digits of interest.

```python
X, y = load_digits(n_class=3, return_X_y=True, as_frame=True)
X = (X>10).astype(int)
```

```python
f, ax = plt.subplots(1,3,figsize=(10,4))
for i in range(3):
    ax[i].imshow(X.iloc[i].values.reshape(8,8))
    ax[i].set_title(f"This is digit {y[i]}.")
plt.suptitle("First three images.")
plt.show()
```

Split data into train and test, let test size be 30% of the dataset and fix random state to 42:

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
assert y_val.shape[0] == 162
assert y_val.sum() == 169
```

Fit a RandomForestClassifier with max_depth=13 and evaluate it's performance:

```python
clf = RandomForestClassifier(max_depth=13, random_state=42)

clf.fit(X_train, y_train)

acc = accuracy_score(y_val, clf.predict(X_val))
print(acc)
```

```python
assert acc > 0.98
```

Now we will use Boruta to find redundand pixels. If the package is not installed in your system, uncomment and run the following cell.

```python
from boruta import BorutaPy

feat_selector = BorutaPy(RandomForestClassifier(max_depth=13), 
                         n_estimators='auto', 
                         verbose=0, 
                         max_iter=100,
                         random_state=42)

feat_selector.fit(np.array(X_train), np.array(y_train))
```

Let us print redundant pixels as a mask. Does the result looks similar to mine (or to Among us chracters)?

At the end let us redo  classification but only with selected features

```python
clf = RandomForestClassifier(max_depth=13, random_state=42)

clf.fit(X_train.loc[:, feat_selector.support_], y_train)

acc = accuracy_score(y_val, clf.predict(X_val.loc[:, feat_selector.support_]))

print(acc)
```

```python
assert acc > 0.99
```

> It seems that RandomForest didn't give any attention to the features deleted by Boruta, so the accuracy is not changed. Or it's already so high - it's hard to do better.
